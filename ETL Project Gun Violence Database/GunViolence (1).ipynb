```python
import pandas as pd
from sqlalchemy import create_engine
import psycopg2
```

```python
#Store Accidental Death CSV into DataFrame
Accidental_Death_Data = "accidental_death_Data.csv"
Accidental_Death_Data_df = pd.read_csv(Accidental_Death_Data)
Accidental_Death_Data_df.head()
```

```python
#Clean Accidental Death DataFrame - This is by removing the columns address and operations
new_accidental_death_df = Accidental_Death_Data_df[['Incident ID', 'Incident Date', 'State', 'City Or County', '# Killed', '# Injured']].copy()
new_accidental_death_df.head()
```

```python
#Rename Columns - rename the columns # killed to Num_killed, #injured to NumInjured
Accidental_Death_Data_df = new_accidental_death_df.rename(columns={"Incident ID":"Incident_ID", "Incident Date":"Incident_Date", "State":"State", "City Or County":"CityorCounty", "# Killed":"Num_Killed", "# Injured":"Num_Injured"})
Accidental_Death_Data_df.head()
```

```python
#Store Accidental Death CSV into DataFrame
Accidental_Injuries = "Accidental_Injuries.csv"
Accidental_Injuries_Data_df = pd.read_csv(Accidental_Injuries)
Accidental_Injuries_Data_df.head()
```

```python
#Clean Accidental Injuries DataFrame - This is by removing the columns address and operations
new_accidental_injuries_df = Accidental_Injuries_Data_df[['Incident ID', 'Incident Date', 'State', 'City Or County', '# Killed', '# Injured']].copy()
new_accidental_injuries_df.head()
```

```python
#Rename Columns - rename the columns # killed to Num_killed, #injured to NumInjured
Accidental_Injuries_Data_df = new_accidental_injuries_df.rename(columns={"Incident ID":"Incident_ID", "Incident Date":"Incident_Date", "State":"State", "City Or County":"CityorCounty", "# Killed":"Num_Killed", "# Injured":"Num_Injured"})
Accidental_Injuries_Data_df.head()
```

```python
#Store Mass Shootings CSV into DataFrame
Mass_Shootings_Data = "Mass_Shootings.csv"
Mass_Shootings_Data_df = pd.read_csv(Mass_Shootings_Data)
Mass_Shootings_Data_df.head()
```

```python
#Clean Mass shootings DataFrame - This is by removing the columns address and operations
new_mass_shootings_df = Mass_Shootings_Data_df[['Incident ID', 'Incident Date', 'State', 'City Or County', '# Killed', '# Injured']].copy()
new_mass_shootings_df.head()
```
